## Qual o valor médio de aluguel por tipo de imóvel?

Aplica apenas o metodo mean() á uma coluna do df:

```python
df.groupby('Tipo')['Valor'].mean()
```

```python
df.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

```python
df_preco_tipo = df.groupby('Tipo')[['Valor']].mean().sort_values('Valor')

df_preco_tipo.plot(kind='barh', figsize=(6,4), color='blue')
```

## Removendo os imóveis comerciais

```python
df.Tipo.unique()
```

```python
imoveis_comerciais = [
    'Conjunto Comercial/Sala',
    'Prédio Inteiro', 'Loja/Salão',
    'Galpão/Depósito/Armazém',
    'Casa Comercial', 'Terreno Padrão',
    'Loja Shopping/ Ct Comercial',
    'Box/Garagem', 'Chácara',
    'Loteamento/Condomínio', 'Sítio',
    'Pousada/Chalé', 'Hotel', 'Indústria'
]
```

```python
df.query('@imoveis_comerciais in Tipo')
```

```python
dados = df.query('@imoveis_comerciais not in Tipo')
dados
```

```python
dados.Tipo.unique()
```

```python
df_preco_tipo = dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')

df_preco_tipo.plot(kind='barh', figsize=(6,4), color='blue')
```
