### Import Pandas 

```python
import pandas as pd
```

### Cria um DF(Data Frame) manualmente

```python
df = pd.DataFrame (
    {
        "Name": [
            "Pedro", 
            "Henrique",
            "Julia"
        ],
        "Age": [23, 24, 22],
        "Sex": ["male", "male", "female"]
    }
)

df
```

### Cada coluna em um *DataFrame* é um número de séries

```python
df["Age"]
```

### OBS: Ao selecionar uma única coluna de um pandas DataFrame, o resultado é um pandas Series. Para selecionar a coluna, use o rótulo da coluna entre colchetes [].

### Series podem ser criados do zero também:

```python
ages = pd.Series([20, 18, 32], name="Age")

ages
```

```python
df["Age"].max()
```

```python
ages.max()
```
